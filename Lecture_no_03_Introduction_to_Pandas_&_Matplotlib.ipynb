**INDEXING METHOD**

iloc and loc are both indexing methods used in pandas to access elements in a DataFrame.


**iloc:**
- Stands for "integer location".
- Uses integer-based indexing to access elements.
- It is primarily used when you want to access DataFrame elements by their integer position.
- The syntax is df.iloc[row_index, column_index].


**loc:**
- Stands for "location".
- Uses label-based indexing to access elements.
- It is primarily used when you want to access DataFrame elements by their index labels or column names.
- The syntax is df.loc[row_label, column_label].

```python
# Accessing via .iloc
# Consider the dataset given in Basic-1.csv. Display second row, second column element.

print(df.iloc[1,1]) #simple indexing
```

```python
print(df.loc[1,'Job Type'])
```

```python
# However, in order to access some rows over single/multiple columns, we can use loc() and iloc() functions as before. The indices will indicate that we are attempting to access multiple rows and/or columns.

# Slicing via .loc
# Access and display first 10 rows from the Gender - Province (both inclusive)


display(df.loc[0:9,'Gender':'Province'])
```

## **matplotlib library**
We mostly use pyplot interface of matplotlib library as it provides a MATLAB-style, procedural, state-machine interface to the underlying object-oriented library in Matplotlib. Pyplot is a collection of command style functions that make Matplotlib work like MATLAB. Each pyplot function makes some change to a figure - e.g., creates a figure, creates a plotting area in a figure etc.

Matplotlib.pyplot is a collection of commands and functions that make Matplotlib behave like MATLAB (for plotting). The MATLAB-style tools are contained in the pyplot (plt) interface.

```python
# Import Matplotlib

import matplotlib.pyplot as plt
```

```python
x1= np.linspace(0,10,100) #100 points linearly spaced between 0 and 10.
plt.plot(x1, np.sin(x1), '-')
plt.show()
```

```python
plt.plot(x1, np.cos(x1), '--')
plt.show()
```

```python
plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')

plt.show()
```

```python
# If we want to create subplots within a figure and not multiple figures, we can divide figures into multiple subplots using the function subplot() within pyplot.

# subplot(nrows, ncols, index)
# divides the figure into nrows and ncols and index refers to which plot you want to access. There are many other arguments. So, for example, the function call subplot(2, 3, 1) which divide the figure into 2 rows and 3 columns (a total of 6 subplots) and the index==1 indicates that the next plot will be plotted in panel 1. (Here indexing starts from 1)

# create a plot figure
plt.figure()


# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))

plt.ylabel('Sine')


# create the second of two panels and set current axis
plt.subplot(212)   # (rows, columns, panel number)
#you can skip the commas and it still works.
plt.plot(x1, np.cos(x1));
plt.ylabel('Cosine')
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x,'r-',label='linear')
plt.plot(x, x**2,'g*', label='quadratic')
plt.plot(x, x**3, 'b^', label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
# Add a new column to dataframe that represents total score. The total score is sum of score of math, reading, and writing scores.

df['Total score']=df['math score']+df['reading score']+df['writing score']
df.describe()
```

```python
df.plot(kind='hist',subplots=True,layout=(1,4),figsize=(20,5))
```

```python
df.hist(figsize=(20,5),layout=(1,4))
```

```python
# Identify non-numeric columns
obj_columns = df.select_dtypes('object').columns

plt.figure(figsize=(20,20)) #resizing the figure
# Plot for each identified column
nrows=ncols=round(len(obj_columns)/2) +1

for i in range(len(obj_columns)): # for the length of columns
        plt.subplot(nrows, ncols,i+1) #for each subplot, out of
        df[obj_columns[i]].value_counts().plot(kind='bar',title=obj_columns[i],rot=45)
plt.show()
```

```python
# Draw pie-chart for the non-numeric columns
#identify non-numeric columns
obj_columns = df.select_dtypes('object').columns

plt.figure(figsize=(20,20)) #resizing the figure
# Plot for each identified column
nrows=ncols=round(len(obj_columns)/2) +1

for i in range(len(obj_columns)): # for the length of columns
        plt.subplot(nrows, ncols,i+1) #for each subplot, out of
        df[obj_columns[i]].value_counts().plot(kind='pie',title=obj_columns[i],rot=45)
plt.show()
```
